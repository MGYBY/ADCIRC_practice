```python
import matplotlib.pyplot as plt
import mikeio
```

```python
filename = "./Global_3D.dfsu"
dfs = mikeio.open(filename)
dfs
```

```python
geom2d = dfs.geometry2d
geom2d
```

```python
dfs.n_layers
```

```python
print(f"Bottom layer: {len(dfs.bottom_elements)}")
```

```python
ds = dfs.read(layers="top")
print(ds)
```

```python
# Set font properties globally
plt.rcParams['font.family'] = "Times New Roman"
# Set font properties globally
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
plt.figure(dpi=600)
plt.rcParams['text.usetex'] = True

ds["Temperature"].plot(vmin=9, vmax=14.0, levels=15, plot_type='shaded', title=False, cmap='Spectral_r', figsize=(11,11));
```
