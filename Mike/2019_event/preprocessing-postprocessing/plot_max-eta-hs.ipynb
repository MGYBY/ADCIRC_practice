```python
import numpy as np
import mikeio
```

## Plot the 2D+SW global $\eta$

```python
import mikeio.generic

fn = "2D_FMSW_modWC_ro/2DFM_SW.mfm"+" "+"-"+" "+"Result Files"+"/global_output.dfsu"
fn_q = "Q_out.dfsu"
mikeio.generic.quantile(fn, fn_q, q=[0.999])
```

```python
ds = mikeio.read(fn_q)
ds
```

```python
import matplotlib.pyplot as plt
# Set font properties globally
plt.rcParams['font.family'] = "Times New Roman"
# Set font properties globally
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
plt.figure(dpi=600)
plt.rcParams['text.usetex'] = True

da_q99 = ds["Quantile 0.999, Surface elevation"]+0.815
da_q99.plot( vmin=0.815, vmax=1.4, levels=19, plot_type='shaded', title=False, cmap='bwr', figsize=(12,9), label='\huge $\zeta_{max}$', save_str="2D_FMSW_modWC_ro/2DFM_SW.mfm"+" "+"-"+" "+"Result Files"+"max_eta.jpeg");
# fig.savefig('./test', dpi=800)
```

```python
ds = mikeio.read("2D_FMSW_modWC_ro/2DFM_SW.mfm"+" "+"-"+" "+"Result Files"+"/global_output.dfsu", time=270)
ds["Surface elevation"]
da = ds.Surface_elevation
da.plot();
```
