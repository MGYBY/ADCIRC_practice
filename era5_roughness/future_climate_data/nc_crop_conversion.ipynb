```python
# year 2099
import os
os.system("rm ./2099_mean_max.txt")
save_file = r'./2099_mean_max.txt'

file_delta = 1
file_begin = 1
file_end = 12

time_count = 0
# 2099 year
for f in np.arange(file_begin, file_end+1, file_delta):
    print(str(f)+': VV')
    if f<10:
        input_file_no_ext = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_VV_2099'+'0'+str(f)
    else:
        input_file_no_ext = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_VV_2099'+str(f)
    input_file = input_file_no_ext+'_mod.nc'
    nc_data = xr.open_dataset(input_file)
    time_count = time_count+len(nc_data.time)
    
# print(time_count)
mean_wind = np.zeros(time_count)
max_wind = np.zeros(time_count)
i_time = 0
for f in np.arange(file_begin, file_end+1, file_delta):
    print(str(f)+': VV')
    if f<10:
        input_file_no_ext_uu = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_UU_2099'+'0'+str(f)
        input_file_no_ext_vv = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_VV_2099'+'0'+str(f)
    else:
        input_file_no_ext_uu = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_UU_2099'+str(f)
        input_file_no_ext_vv = './2099/'+'NEMO_GL_MPI-ESM1-2-HR_SSP5-RCP8.5_324x200-cfl3_VV_2099'+str(f)
    input_file_uu = input_file_no_ext_uu+'_mod.nc'
    input_file_vv = input_file_no_ext_vv+'_mod.nc'
    nc_data_uu = xr.open_dataset(input_file_uu)
    nc_data_vv = xr.open_dataset(input_file_vv)
    vmag = ((nc_data_uu.UU.values)**2.0+(nc_data_vv.VV.values)**2.0)**0.5
    # TODO: use Pandas timeframe
    for j in np.arange(len(nc_data_uu.time)):
        mean_wind[i_time] = np.mean(vmag[j])
        max_wind[i_time] = np.max(vmag[j])
        i_time = i_time+1

import csv
file_name = '2099_mean_max.txt'
hours = np.arange(1, time_count+1)
with open(file_name, 'w') as f:
    writer = csv.writer(f, delimiter=' ')
    writer.writerows(zip(np.transpose(hours), np.transpose(mean_wind), np.transpose(max_wind)))
```

```python
print(len(mean_wind))
fig, ax = plt.subplots(figsize = (12.5, 6))
ax.plot(np.arange(1, time_count+1), mean_wind, linewidth = 1.5)
```
